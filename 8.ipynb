```python
def string_to_num_list(d):
    data = d.split(' ')
    return list(map(lambda x: int(x), data))
```

```python
def get_node_meta_sum(d, i):
    node_count = d[i]
    meta_count = d[i + 1]
    
    if node_count == 0:
        return sum(d[i + 2: i + meta_count + 2]), meta_count + 2
    
    child_sums = []
    cum_size = 0
    for j in range(node_count):
        acc, node_size = get_node_meta_sum(d, i + cum_size + 2)
        
        child_sums.append(acc)
        cum_size += node_size
        
    return sum(child_sums) + sum(d[i + 2 + cum_size: i + 2 + cum_size + meta_count]), cum_size + meta_count + 2
```

```python
# part 2
def get_node_value(d, i):
    node_count = d[i]
    meta_count = d[i + 1]
    
    if node_count == 0:
        return sum(d[i + 2: i + meta_count + 2]), meta_count + 2
    
    cum_size = 0
    child_values = []
    for j in range(node_count):
        value, node_size = get_node_value(d, i + cum_size + 2)
        child_values.append(value)
        cum_size += node_size
    
    node_acc = 0
    nodes_to_take = d[i + 2 + cum_size: i + 2 + cum_size + meta_count]
    print(nodes_to_take)
    for j in nodes_to_take:
        if j  < len(child_values):
            node_acc += child_values[j]
    
    return node_acc, cum_size + meta_count + 2
```
