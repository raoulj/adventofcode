```python
# Setup work 

# Get the data
import requests
r = requests.get('https://adventofcode.com/2018/day/3/input', cookies={
    'session': '53616c7465645f5fa2ecece720654838baaaf96bc125b8893c564f010d5ad7a90b3dab0760aef0074c4a7ec162320c3e'
})
```

```python
claims = list(filter(len,r.text.split('\n')))
```

```python
def raw_claim_to_structured_claim(s):
    without_space = s.replace(' ', '')
    without_id = without_space.split('@')[1]
    coords_and_size = without_id.split(':')
    coords = coords_and_size[0].split(',')
    size = coords_and_size[1].split('x')
    return list(map(int, coords + size))
    
structured_claims = list(map(raw_claim_to_structured_claim, claims))
```

```python
import numpy as np
grid = np.zeros((1500, 1500))

for claim in structured_claims:
    for x in range(claim[0],claim[0]+ claim[2]):
        for y in range(claim[1],claim[1]+ claim[3]):
            grid[x][y] += 1
```

```python
acc = 0
for x in range(len(grid)):
    for y in range(len(grid[0])):
        if grid[x][y] > 1:
            acc += 1
```

```python
# part 2
def check_claim(c):
    for x in range(c[0],c[0]+ c[2]):
        for y in range(c[1],c[1]+ c[3]):
            if grid[x][y] > 1:
                return
    print(c)
for claim in structured_claims:
    check_claim(claim)
    
```
